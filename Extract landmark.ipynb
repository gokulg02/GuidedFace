```python
import cv2
import dlib
import numpy as np
import os

# Facial landmark indices
FACIAL_LANDMARKS_INDEXES = {
    "jaw": (0, 17),
    "right_eyebrow": (17, 22),
    "left_eyebrow": (22, 27),
    "nose_bridge": (27, 31),
    "lower_nose": (31, 36),
    "right_eye": (36, 42),
    "left_eye": (42, 48),
    "outer_lip": (48, 60),
    "inner_lip": (60, 68)
}

# Directories
input_dir = "./Data/RestructurePrompt/target"
output_dir = "./Data/RestructurePrompt/source"

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Load detector and predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("./Data/RestructurePrompt/shape_predictor_68_face_landmarks.dat")

# Supported image formats
valid_exts = [".jpg", ".jpeg", ".png", ".bmp"]

flag_i=0
# Process all images
for filename in os.listdir(input_dir):
    if not any(filename.lower().endswith(ext) for ext in valid_exts):
        continue  # skip non-image files

    input_image_path = os.path.join(input_dir, filename)
    output_image_path = os.path.join(output_dir, filename)

    # Load and process image
    image = cv2.imread(input_image_path)
    if image is None:
        print(f"Warning: Could not load image {input_image_path}")
        continue

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    landmark_img = np.zeros_like(image)

    # Detect faces
    faces = detector(gray)
    if not faces:
        # print(f"No faces found in {filename}")
        no_face.append(filename)
        continue

    for face in faces:
        landmarks = predictor(gray, face)

        for name, (start, end) in FACIAL_LANDMARKS_INDEXES.items():
            points = []
            for i in range(start, end):
                x, y = landmarks.part(i).x, landmarks.part(i).y
                points.append((x, y))
                # Draw dot
                # cv2.circle(landmark_img, (x, y), 5, (255, 255, 255), -1)

            is_closed = name in ["left_eye", "right_eye", "outer_lip", "inner_lip"]
            for j in range(len(points) - 1):
                cv2.line(landmark_img, points[j], points[j + 1], (255, 255, 255), 5)
            if is_closed:
                cv2.line(landmark_img, points[-1], points[0], (255, 255, 255), 5)

    # Save image
    cv2.imwrite(output_image_path, landmark_img)
    print(flag_i)
    flag_i+=1
    # print(f"Saved: {output_image_path}")
```

```python
import os
from PIL import Image

input_dir = "./Data/RestructurePrompt/source"  # change this to your directory
invalid_images = []

for filename in os.listdir(input_dir):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
        path = os.path.join(input_dir, filename)
        try:
            with Image.open(path) as img:
                if img.size != (512, 512):
                    print(f"Invalid size: {filename} - {img.size}")
                    invalid_images.append(filename)
        except Exception as e:
            print(f"Error reading {filename}: {e}")

if not invalid_images:
    print("All images are 512x512.")
else:
    print(f"{len(invalid_images)} image(s) are not 512x512.")
```
